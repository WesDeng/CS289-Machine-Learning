```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

**3. Isocountours of Normal Distribution**

```python
def plot_contour(mean, cov):
    a, b = mean[0]-4, mean[1]+4   
    x, y = np.mgrid[a:b:.01, a:b:.01] # Choose a fixed location.
    pos = np.dstack((x, y))
    rv = multivariate_normal(mean, cov)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.contourf(x, y, rv.pdf(pos))
```

1.

```python
mean = [1, 1]
cov = [[1, 0],
       [0, 2]]
plot_contour(mean, cov)
```

2.

```python
mean = [-1, 2]
cov = [[2, 1],
       [1, 4]]
plot_contour(mean, cov)
```

```python
def plot_subtraction_contour(mean_1, mean_2, cov_1, cov_2):
    x, y = np.mgrid[-3.5:3.5:.01, -3.5:3.5:.01]
    pos = np.dstack((x, y))
    rv_1 = multivariate_normal(mean_1, cov_1)
    rv_2 = multivariate_normal(mean_2, cov_2)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.contourf(x, y, rv_1.pdf(pos) - rv_2.pdf(pos))
```

3.

```python
mean_1, mean_2 = [0, 2], [2, 0]
cov_1 = [[2, 1],
         [1, 1]]
cov_2 = cov_1

plot_subtraction_contour(mean_1, mean_2, cov_1, cov_2)
```

4.

```python
mean_1, mean_2 = [0, 2], [2, 0]
cov_1 = [[2, 1],
         [1, 1]]
cov_2 = [[2, 1],
         [1, 4]]
plot_subtraction_contour(mean_1, mean_2, cov_1, cov_2)
```

5.

```python
mean_1, mean_2 = [1, 1], [-1, -1]
cov_1 = [[2, 0],
         [0, 1]]
cov_2 = [[2, 1],
         [1, 2]]
plot_subtraction_contour(mean_1, mean_2, cov_1, cov_2)
```
